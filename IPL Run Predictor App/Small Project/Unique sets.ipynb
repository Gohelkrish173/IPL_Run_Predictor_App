```python
import pandas as pd

# Load the CSV file to inspect its contents
file_path = 'deliveries.csv'
df = pd.read_csv(file_path)

# Find unique values for batting_team, batter, and bowler
unique_batting_teams = list(set(df['batting_team']))
unique_batters = list(set(df['batter']))
unique_bowlers = list(set(df['bowler']))

# Show the length of unique items in each list and display a few examples
len(unique_batting_teams), len(unique_batters), len(unique_bowlers), unique_batting_teams, unique_batters, unique_bowlers
```
